```python
from matplotlib import pyplot as mp
import numpy as np
```

<img src="gaussian formula.png"/>

```python
pi = np.pi
print (pi)
```

```python
def gaussian(x,mu,sigma):
    A = sigma * np.sqrt(2.*pi)
    B = -0.5*np.power((x-mu)/sigma,2.)
    return (1/A)*np.exp(B)
```

```python
def stdNormal(x):
    return gaussian(x, 0.,1.)
```

```python
stdNormal(0.), stdNormal(1.), stdNormal(2.)
```

```python
stdNormal(0.), stdNormal(-1.), stdNormal(-2.)
```

```python
xValues = np.linspace(-5.,5.,11)
xValues
```

```python
yValues = stdNormal(xValues)
yValues
```

```python
mp.plot(xValues, yValues)
```

```python
xValues = np.linspace(-5,5,100)
yValues = stdNormal(xValues)
plots = mp.plot(xValues,yValues)
```

```python
for mu, sigma in [(0.,1.), (0.,2.), (1.,0.5), (1.,0.75)]:
    yValues = gaussian(xValues, mu, sigma)
    mp.plot(xValues, yValues)
```

```python
for mu, sigma in [(0.,1.), (0.,2.), (1.,0.5), (1.,0.75)]:
    yValues = gaussian(xValues, mu, sigma)
    mp.plot(xValues, yValues, label = "mu={} and sigma={}".format(mu, sigma))
    mp.legend(loc="upper left", ncol=1, title = "Several Gaussian Plots")
```
